```python
df_men_usa = df[['Date','Gender','Country', 'Size (US)']].copy()
df_men_usa
```

```python
df_men_usa = df_men_usa[df_men_usa.Gender != 'Female']
df_men_usa = df_men_usa[df_men_usa.Country == 'United States']
df_men_usa = df_men_usa[df_men_usa.Date >= '2015-01']
df_men_usa = df_men_usa[df_men_usa.Date <= '2016-12']

df_men_usa.head(660)
```

```python
df_groupby_size_date = df_men_usa.groupby(['Date', 'Size (US)']).agg({'Gender': pd.Series.count })
df_groupby_size_date.head(20)
```

```python
df_groupby_size_date_2 = df_groupby_size_date.groupby(['Size (US)']).agg({'Gender': pd.Series.sum }) 
meanX = df_groupby_size_date_2 / 24
meanX.round(2)
```

```python
pre_variance = (df_groupby_size_date - meanX)**2
variance  = pre_variance.groupby('Size (US)').agg({'Gender': pd.Series.sum}) / 24 
variance
```

```python
std = variance.Gender ** 0.5
std
```

```python
std_error = std / math.sqrt(24)
std_error
```

COFIDENCE INTERVALS

```python
T = 2.64
ci_a = meanX.Gender + (T * std_error) 
ci_a.round(2)
```

```python
ci_b = meanX.Gender - (T * std_error) 
ci_b.round(2)
```
